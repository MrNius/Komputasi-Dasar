```python
## 5.7 (Duplicate Elimination)

def eliminasi_duplikat(list):       
    list.sort()      
    list_unik = []      
    for i in list:         
        if i not in list_unik:             
            list_unik.append(i)     
    return list_unik         

list_angka = [2, 56, 2, 34, -1, -34, -789, 56, 1, -56, 1, 2, 34, 56, 789, -1, -2, -34, -56, 0, -789]
list_kata = ['donat', 'Bagas', 'apel', 'Agus', 'biskuit', 'Candra', 'Bagas', 'cendol', 'Danu', 'Candra', 'donat', 'apel', 'Danu']
print('Angka unik terurut :', eliminasi_duplikat(list_angka)) 
print('Kata unik terurut :', eliminasi_duplikat(list_kata))
```

```python
## 5.8 (Sieve of Eratosthenes)

import math
bilangan = [True for i in range(1000)]
bilangan[0] = bilangan[1] = False # 0 dan 1 bukan bilangan prima

for i in range(2, round(math.sqrt(999))):
    if(bilangan[i] == True):
        j = i*i
        while j < 1000: # membuat kelipatan dari bilangan prima i menjadi False
            bilangan[j] = False
            j += i
            
# menampilkan bilangan prima (bilangan yang True)
print("Bilangan prima kurang dari 1000 :")
for prima in range(1000):
    if bilangan[prima] == True:
        print(prima, end='\t')
```

```python
## 5.9 (Palindrome Tester)

import string
def is_palindrome(kata):
    def membersihkan_kata(kata):
        kata = kata.lower()
        kata = ''.join(c for c in kata if c not in string.punctuation and not c.isspace())
        return kata
    kata = membersihkan_kata(kata)
    tumpukan = []
    for i in kata:
        tumpukan.append(i)
    kata_terbalik =''.join(tumpukan[::-1])
    return kata == kata_terbalik

print('Menggunakan fungsi is_palindrome untuk mengecek apakah kata-kata berikut merupakan palindrom atau bukan :')
print('1. "Level" :', is_palindrome("Level"))
print('2. "world" :', is_palindrome("world"))
print('3. "race car" :', is_palindrome("race car"))
print('4. "(A man, a plan, a canal, Panama)" :', is_palindrome("A man, a plan, a canal, Panama"))
```

```python
## 5.10 (Anagrams)

import itertools 
def anagrams(kata):     
    hasil = ["".join(i) for i in itertools.permutations(kata)]     
    print(list(set(hasil))) # Menggunakan fungsi set untuk menghilangkan duplikat, bentuk dalam list

# Contoh penggunaan fungsi
print('Anagram dari kata "dan" :')
anagrams("dan")
print('Anagram dari kata "yang" :')
anagrams("yang")
print('Anagram dari kata "jatuh" :')
anagrams("jatuh")
```

```python
## 5.11 (Summarizing Letters in a String)

import string
def summarize_letters(kalimat):
    def bersih(kalimat): # Membersihkan kalimat dari spasi dan tanda baca
        kalimat = kalimat.lower()
        kalimat = ''.join(c for c in kalimat if c not in string.punctuation and not c.isspace())
        return kalimat

    kalimat = bersih(kalimat) # Kalimat bersih
    hitung_huruf = {}

    for i in kalimat:
        if i.isalpha(): # Periksa apakah karakternya berupa huruf
            if i in hitung_huruf:
                hitung_huruf[i] += 1
            else:
                hitung_huruf[i] = 1

    # Periksa apakah semua huruf alfabet ada
    alfabet = set(string.ascii_lowercase)
    alfabet_lengkap = set(hitung_huruf.keys()) == alfabet

    return list(hitung_huruf.items()), alfabet_lengkap

# Menguji fungsi dengan kalimat masukan
## Kalimat "The quick brown fox jumps over the lazy dog" mengandung semua huruf alfabet di dalamnya.
kalimat = input('Masukan kalimat untuk dicek frekuensi hurufnya :')
frekuensi_huruf, alfabet_lengkap = summarize_letters(kalimat)

# Menampilkan frekuensi kata
for huruf, frekuensi in frekuensi_huruf:
    print(f" Huruf '{huruf}' muncul {frekuensi} kali")

# Periksa apakah kalimat memiliki semua huruf alfabet
if alfabet_lengkap == True:
    print("Kalimat tersebut memiliki semua huruf alfabet.")
else:
    print("Kalimat tersebut tidak memiliki semua huruf alfabet.")
```

```python
## 5.12 (Telephone-Number Word Generator)

korespondensi = {
    2:['A', 'B', 'C'], 
    3:['D', 'E', 'F'], 
    4:['G', 'H', 'I'], 
    5:['J', 'K', 'L'], 
    6:['M', 'N', 'O'], 
    7:['P', 'R', 'S'], 
    8:['T', 'U', 'V'], 
    9:['W', 'X', 'Y']
}

def menghasilkan_kata(nomor, kata=['']):
    if nomor == 0:
        return kata
    kata_baru = []
    kosong = nomor % 10
    if kosong == 0 or kosong == 1: # Memastikan nomor telpon tidak mengandung 0 atau 1
        return ['Tidak ada']
        
    for i in korespondensi.get(kosong, []):
        for j in kata:
            kata_baru.append(i+j) 
    return menghasilkan_kata(nomor//10, kata_baru)

nomor = int(input('Masukan nomor telpon: ')) 
# Contoh : '7382273' untuk 'PETCARE', '6862377' untuk 'NUMBERS', '2443282' untuk 'BIGDATA'
print('Kemungkinan kombinasi kata yang sesuai dengan nomor tersebut :')
for i in menghasilkan_kata(nomor):
    print(i)
```

```python
## 5.15 (Tuples Representing Invoices)

from operator import itemgetter

# Tuple data invoices (part ID, part description, quantity, item price)
invoices = [
    ("83", "Electric sander", 7, 57.98),
    ("24", "Power saw", 18, 99.99),
    ("7", "Sledge hammer", 11, 21.50),
    ("77", "Hammer", 76, 11.99),
    ("39", "Jig saw", 3, 79.50)
]

# a) Mengurutkan berdasarkan part description
sort_description = sorted(invoices, key=itemgetter(1))
print("a) Mengurutkan berdasarkan part description :")
for invoice in sort_description:
    print(invoice)

# b) Mengurutkan berdasarkan price
sort_price = sorted(invoices, key=itemgetter(3))
print("\nb) Mengurutkan berdasarkan price :")
for invoice in sort_price:
    print(invoice)

# c) Mengurutkan berdasarkan quantity
sort_quantity = sorted([(invoice[1], invoice[2]) for invoice in invoices], key=itemgetter(1))
print("\nc) Mengurutkan berdasarkan quantity :")
for part_description, quantity in sort_quantity:
    print(f"{part_description}: {quantity}")

# d) Mengurutkan berdasarkan invoice value
sort_invoice_value = sorted([(invoice[1], invoice[2] * invoice[3]) for invoice in invoices], key=itemgetter(1))
print("\nd) Mengurutkan berdasarkan invoice value :")
for part_description, invoice_value in sort_invoice_value:
    print(f"{part_description}: ${invoice_value:.2f}")

# e) Filter invoice values dalam range $200 - $500
filtered_invoices = [(invoice[1], invoice[2] * invoice[3]) for invoice in invoices if 200 <= invoice[2] * invoice[3] <= 500]
print("\ne) Invoice values dalam range $200 - $500 :")
for part_description, invoice_value in filtered_invoices:
    print(f"{part_description}: ${invoice_value:.2f}")

# f) Menghitung total semua invoice
total_invoices = sum(invoice[2] * invoice[3] for invoice in invoices)
print("\nf) Total semua invoice : ${:.2f}".format(total_invoices))
```

```python
## 5.17 (Filter/Map Performance) 

numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

def filtering(x):
    print(f"Filtering {x}")
    return x % 2 != 0

def mapping(x):
    print(f"Mapping {x}")
    return x ** 2

# a)
angka_terfilter = list(filter(filtering, numbers))
print('a) Berapa kali filtering memanggil argumen lambdanya?', len(angka_terfilter), 'kali')
print(f"> Angka ganjil yang terfilter : {angka_terfilter} \n")

# b)
angka_kuadrat = list(map(mapping, angka_terfilter))
print('b) Berapa kali mapping memanggil argumen lambdanya sesudah difilter?', len(angka_kuadrat), 'kali')
print(f"> Angka kuadrat hasil pemetaan : {angka_kuadrat} \n")

# c)
angka_kuadrat_terbalik = list(map(mapping, numbers))
angka_terfilter_terbalik = list(filter(filtering, angka_kuadrat_terbalik))
print('c) Berapa kali mapping memanggil argumen lambdanya sebelum difilter (reversed)?', len(angka_kuadrat_terbalik), 'kali')
print(f"> Angka kuadrat hasil pemetaan sebelum difilter (reversed): {angka_kuadrat_terbalik}")
```

```python
## 6.5 (Counting Duplicate Words)

text = ('Ini adalah contoh Kalimat yang memiliki kata duplikat Misalnya Kata Kata Kata Contoh kalimat ini juga Memiliki aturan penulisan Yang salah')
text_new = text.lower()
word_counts = {}

# count occurrences of each unique word
for word in text_new.split():
    if word in word_counts:
        word_counts[word] += 1 #increment count
    else:
        word_counts[word] = 1
        
print(f'{"KATA":<19} JUMLAH')

for word, count in sorted(word_counts.items()):
        print(f'{word:<20}{count}')
    
print('\nJumlah kata-kata unik:', len(word_counts) )

print("Kata-kata duplikat:")
for word, count in word_counts.items():
    if word_counts[word] > 1:
        print(f' {word:<20}')
```

```python
## 6.6 (Duplicate Word Removal)

def unique_words(sentence_list):
    unique_word_set = set()  # Set untuk menyimpan kata-kata unik
    for sentence in sentence_list:
        words = sentence.lower().split()  # Mengonversi kata-kata menjadi huruf kecil dan memecah kalimat menjadi kata-kata
        unique_word_set.update(words)  # Menambahkan kata-kata ke set
    
    unique_word_list = sorted(unique_word_set)  # Mengurutkan kata-kata unik dalam urutan alfabet
    return unique_word_list

# Uji fungsi dengan beberapa kalimat
sentences1 = ["Ini adalah contoh kalimat pertama yang terdapat kata duplikat misalnya kata kata kata"]
sentences2 = ["Contoh kalimat kedua adalah ini ada kata duplikat juga seperti ada ada ada"]
sentences3 = ["Kalimat ketiga juga merupakan contoh yang memiliki kata duplikat contohnya juga juga juga"]
result1 = unique_words(sentences1)
result2 = unique_words(sentences2)
result3 = unique_words(sentences3)

print(f'\nKata-kata unik dalam kalimat "{sentences1[0]}" adalah :')
for word in result1:
    print('', word)
print(f'\nKata-kata unik dalam kalimat "{sentences2[0]}" adalah :')
for word in result2:
    print('', word)
print(f'\nKata-kata unik dalam kalimat "{sentences3[0]}" adalah :')
for word in result3:
    print('', word)
```

```python
## 6.7 (Character Counts)

text = input("Masukan sebuah kalimat:").lower()
# creating an empty dict
char_count = dict()
# looping through each char
for c in text:
    # checking if c is not a space
    if c != " ":
        # if c is already in char_count, incrementing count by 1
        if c in char_count:
            char_count[c] += 1
        # otherwise adding to char_count with key=c, value=1
        else:
            char_count[c] = 1
print('HURUF\t\tJUMLAH')
# looping and printing each letter and its count in the sorted order of letters
for c in sorted(char_count.keys()):
    print('{}\t\t{}'.format(c, char_count[c]))
# part B, creating a set of all alphabets
alphabets = set('abcdefghijklmnopqrstuvwxyz')
# using difference method of set, finding characters in alphabets that are not in char_count dict
alphabets_not_in_input = alphabets.difference(char_count.keys())
# displaying the absent characters in sorted order
print("\nHuruf yang tidak terdapat dalam kalimat yang dimasukan:", sorted(alphabets_not_in_input))
```

```python
## 6.8 (Challenge: Writing the Word Equivalent of a Check Amount) 

def angka_jadi_kata(angka):
    kata = { 
        0 : 'zero', 
        1 : 'one', 
        2 : 'two', 
        3 : 'three', 
        4 : 'four', 
        5 : 'five',
        6 : 'six',
        7 : 'seven', 
        8 : 'eight', 
        9 : 'nine', 
        10 : 'ten',
        11 : 'eleven', 
        12 : 'twelve', 
        13 : 'thirteen', 
        14 : 'fourteen',
        15 : 'fifteen', 
        16 : 'sixteen', 
        17 : 'seventeen', 
        18 : 'eighteen',
        19 : 'nineteen', 
        20 : 'twenty',
        30 : 'thirty', 
        40 : 'forty', 
        50 : 'fifty', 
        60 : 'sixty',
        70 : 'seventy', 
        80 : 'eighty', 
        90 : 'ninety' 
    }
    
    ribu = 1000
    juta = ribu * 1000

    assert(0 <= angka)

    if (angka < 20):
        return kata[angka]

    elif (angka < 100):
        if angka % 10 == 0: 
            return kata[angka]
        else: 
            return kata[angka // 10 * 10] + ' ' + kata[angka % 10]

    elif (angka < ribu):
        if angka % 100 == 0: 
            return kata[angka // 100] + ' hundred'
        else: 
            return kata[angka // 100] + ' hundred ' + angka_jadi_kata(angka % 100)

    elif (angka < juta):
        if angka % ribu == 0: 
            return angka_jadi_kata(angka // ribu) + ' thousand'
        else: 
            return angka_jadi_kata(angka // ribu) + ' thousand, ' + angka_jadi_kata(angka % ribu)
      
jumlah = input('Masukan angka yang ingin dijadikan kata:')
pecah = jumlah.split('.')
if len(pecah) == 2 :
   print(angka_jadi_kata(int(pecah[0])).upper(), "AND", pecah[1], "/", 10**(len(pecah[1])))
else :
   print(angka_jadi_kata(int(pecah[0])).upper())
```
